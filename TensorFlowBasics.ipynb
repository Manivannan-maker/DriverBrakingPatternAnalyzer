# Basic ML in tensor Flow

This tutorial will give basic introduction about ML usage in Tensorflow 

Steps to be followed in Creating a Machine Learning from scratch

1.   Import dependencies
2.   Building Neural Network
3.   Training Data Generation
4.   Train the neural network
5.   Run the Machine learning Model






# 1. Import dependencies
Before proceeding with programming , we need to install basic dependencies like tensorflow package , graphical library etc

```python
#import dependencies 
import tensorflow as tf
import numpy as np
from tensorflow import keras
# Matplotlib is a graphing library
import matplotlib.pyplot as plt
# Math is Python's math library
import math
```

# 2. Building Neural network
In this example, you know that the relationship between the numbers is Y=2X+1.
Now we are going to create a machine learning model to predict this relationship.

Next, create the simplest possible neural network. It has one layer, that layer has one neuron, and the input shape to it is only one value.





```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```


When the computer is trying to learn that, it makes a guess, maybe Y=10X+10. The loss function measures the guessed answers against the known correct answers and measures how well or badly it did.

Next, the model uses the optimizer function to make another guess. Based on the loss function's result, it tries to minimize the loss. At this point, maybe it will come up with something like Y=5X+5. While that's still pretty bad, it's closer to the correct result (the loss is lower).

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# 3. Training Data Generation
We are going to generate datasets as  x and corresponding Y values to train the model

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-1.0, 1.0, 3.0, 6.0, 7.0, 9.0], dtype=float)
```

Plot the value in graphical view

```python
# Plot our data. The 'b.' argument tells the library to print blue dots.
plt.plot(xs, ys, 'b.')
plt.show()
```

# 4. Train the neural network
The process of neural network is to train the relationship between X and Y values in the given dataset .
model.fit will be called for model training , where it runs 'n' number of epochs in the loop to find the how good or bad the expected result came by comparing the corresponding Y given in datasets . The optimizer will make closer guess based on the previous epoch's loss / deviations from output.


```python
model.fit(xs, ys, epochs=300)
```

# Run the Machine learning Model 
Give the X value in model.predict , The output will be closer to Y=2X+1 formula.



```python
print(model.predict([10.0]))
```
